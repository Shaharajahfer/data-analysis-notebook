# Crime Data Analysis (Los Angeles 2020–2024)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
crime = pd.read_csv('Crime_Data_from_2020_to_2024_20260414.csv')
```

## Understand the basic structure of the data, including data types and missing values.

```python
crime.head(10)
```

## Data Cleaning

```python
crime.isna().sum()
```

```python
df = crime.drop(labels=['Mocodes', 'Weapon Used Cd', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street'], axis = 1
               )
df
```

```python
df.sort_values(by = 'Date Rptd')
```

```python
df.dtypes
```

## Update datatypes

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'],format = '%Y %b %d %I:%M:%S %p')
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'],format = '%Y %b %d %I:%M:%S %p')
df
```

```python
df.info()
```

```python
df.describe(include = ['int64', 'float64'])
```

```python
df['Vict Age'].value_counts()
```

```python
# Age should be greater than zero
df_age = df[df['Vict Age'] > 0]
df_age['Vict Age'].value_counts()
```

### Crime Trends Over Time 

```python
df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month
df
```

```python
df['Month'].value_counts()
```

### Line chart shows number of crimes in each month

```python
df['Month'].value_counts().sort_index().plot().set_ylabel('No of crimes')
```

```python
df['Year'].value_counts().sort_index()
```

### Bar chart shows number of crimes in each year

```python
df['Year'].value_counts().sort_index().plot(kind = 'bar').set_ylabel('No of crimes')
```

```python
df.groupby(['Year','Month']).size()
```

### This line chart shows the number of crimes in each combination of month and year.

```python
df.groupby(['Year','Month']).size().plot().set_ylabel('No of crimes')
```

### The histogram shows the age distribution of criminals

```python
plt.hist(df_age['Vict Age'], edgecolor = 'black', bins = 15)
plt.xlabel('Age')
plt.show()
```

### Crime count per day

```python
crime_count_per_day = df.groupby(['DATE OCC', 'Crm Cd Desc']).size()
crime_count_per_day
```

### Top 20 crimes 

```python
top_20_crimes = crime_count_per_day.groupby('Crm Cd Desc').mean().reset_index(name = 'average_daily_count')\
                 .sort_values(by = 'average_daily_count', ascending = False).head(20)
top_20_crimes
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by = 'average_daily_count')
top_20_crimes_sorted
```

### Horizontal bar chart of top 20 crimes

```python
plt.barh(top_20_crimes_sorted['Crm Cd Desc'], top_20_crimes_sorted['average_daily_count'])
```

### Formatting the plots

```python
plt.style.available
plt.style.use('ggplot')
```

```python
plt.figure(figsize = (12, 8))
# Line chart
plt.subplot(2,2,1)
crime_plot = df.groupby(['Year','Month']).size().plot(color = 'green', linewidth = 2)
crime_plot.set_xlabel('Year-Month', fontsize = 10)
crime_plot.set_title('Crime Over Time')

# Histogram
plt.subplot(2,2,2)
plt.hist(df_age['Vict Age'], edgecolor = 'black', bins = 15)
plt.xlabel('Age', fontsize = 10)
plt.title('Distribution of Victim Age')
plt.tight_layout()

# Horizontal bar chart
plt.figure()
plt.barh(top_20_crimes_sorted['Crm Cd Desc'], top_20_crimes_sorted['average_daily_count'], color = 'orange')
plt.title('Top 20 Crimes')
```

### The time difference between crimes that occurred and those that were reported.

```python
df['Time to Rept'] = (pd.to_datetime(df['Date Rptd']) - pd.to_datetime(df['DATE OCC'])).dt.days
```

```python
df
```

```python
df['Time to Rept'].value_counts()
```

```python
df.groupby(['Crm Cd Desc'])['Time to Rept'].value_counts()
```

### Average time taken to report each crime

### The areas in which highest number of crimes occuring during night time

```python
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'].astype(str).str.zfill(4), format='%H%M').dt.time
```

```python
nighttime_crimes = df[(df['TIME OCC'].apply(lambda t: t.hour) >= 20)\
    | (df['TIME OCC'].apply(lambda t: t.hour) <= 3)]
nighttime_crimes 
```

```python
nighttime_crimes['AREA NAME'].value_counts()
```

### Hourly occurance of crimes

```python
hourly_occ = df['TIME OCC'].apply(lambda t: t.hour).value_counts().sort_index().reset_index(name = 'count')
hourly_occ
```

### Bar chart shows the hourly occurance of crimes

```python
plt.bar(hourly_occ['TIME OCC'], hourly_occ['count'])
plt.xlabel('Hour')
plt.ylabel('No of crimes')
```

```python
# Maximum number of crimes occuring at 12 o’clock
# Count of each crime occurring at 12 o’clock.
df[df['TIME OCC'].apply(lambda t: t.hour) == 12]['Crm Cd Desc'].value_counts()
```

```python
#The crime that occurs most often at 12 o’clock is theft of identity.
no_identity_theft = df[~(df['Crm Cd Desc'] == 'THEFT OF IDENTITY')]
```

### Hourly occurance of crime without theft of identity

```python
new_hourly_occ = no_identity_theft['TIME OCC'].apply(lambda t: t.hour).value_counts().sort_index().reset_index(name = 'count')
```

```python
plt.bar(new_hourly_occ['TIME OCC'], new_hourly_occ['count'])
plt.xlabel('Hour')
plt.ylabel('No of crimes')
```
